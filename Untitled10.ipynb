```python
mean_price = df['YearStart'].mean()
median_price = df['YearStart'].median()
std_dev_price = df['YearStart'].std()

print(f"Mean 	YearStart: {mean_price:.4f}")
print(f"Median 	YearStart: {median_price:.4f}")
print(f"Standard Deviation of 	YearStart : {std_dev_price:.4f}")
```

```python
Q1 = df['YearStart'].quantile(0.25)
print(Q1)
```

```python
Q3 = df['YearStart'].quantile(0.75)
print(Q3)
```

```python
IQR = Q3 - Q1
```

```python
lower_bound = Q1 - 1.5 * IQR
print(lower_bound)
```

```python
upper_bound = Q3 + 1.5 * IQR
print(upper_bound)
```
